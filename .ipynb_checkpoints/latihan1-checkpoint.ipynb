```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = pd.read_csv('data.csv')
print(data.tail())
```

```python


medianPm=data['PM2.5'].median()
data['PM2.5'].fillna(medianPm)

medianPm10=data['PM10'].median()
data['PM10'].fillna(medianPm10)
medianSO2=data['SO2'].median()
data['SO2'].fillna(medianSO2)
medianN02=data['NO2'].median()
data['NO2'].fillna(medianN02)
medianCO=data['CO'].median()
data['CO'].fillna(medianCO)
median03=data['O3'].median()
data['O3'].fillna(median03)
data=data.dropna(axis=0)
```

```python
data.isna().sum()
```

```python
bulan_mapping = {
    1: "Januari", 2: "Februari", 3: "Maret", 4: "April",
    5: "Mei", 6: "Juni", 7: "Juli", 8: "Agustus",
    9: "September", 10: "Oktober", 11: "November", 12: "Desember"
}
data['month'] = data['month'].replace(bulan_mapping)
```

```python
data.head()
```

```python
ganti_nama_arah_angin = {
    'N': 'North',
    'NNE': 'North-North-East',
    'NE': 'North-East',
    'ENE': 'East-North-East',
    'E': 'East',
    'ESE': 'East-South-East',
    'SE': 'South-East',
    'SSE': 'South-South-East',
    'S': 'South',
    'SSW': 'South-South-West',
    'SW': 'South-West',
    'WSW': 'West-South-West',
    'W': 'West',
    'WNW': 'West-North-West',
    'NW': 'North-West',
    'NNW': 'North-North-West'
}

# Terapkan mapping ke kolom 'wd'
data['wd'] = data['wd'].replace(ganti_nama_arah_angin )
```

```python
data.head()
```

```python
koloms = ['PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

for kolom in koloms:
    q1 = data[kolom].quantile(0.25)
    q3 = data[kolom].quantile(0.75)
    iqr = q3 - q1
    batas_bawah = q1 - 1.5 * iqr
    batas_atas = q3 + 1.5 * iqr
    
    data[kolom] = np.where(data[kolom] < batas_bawah, batas_bawah, 
                          np.where(data[kolom] > batas_atas, batas_atas, data[kolom]))
```

```python



def classify_air_quality(row):
    if (row['PM2.5'] <= 12) and (row['PM10'] <= 50) and (row['SO2'] <= 20) and (row['NO2'] <= 40) and (row['CO'] <= 1) and (row['O3'] <= 100):
        return ' Baik'
   
    elif (row['PM2.5'] <= 35.4) and (row['PM10'] <= 100) and (81 <= row['SO2'] <= 80) and (row['NO2'] <= 200) and ( row['CO'] <= 3) and (row['O3'] <= 180):
        return 'Sedang'
    elif (row['PM2.5'] <= 55.4) and (row['PM10'] <= 150) and (381 <= row['SO2'] <= 380) or (row['NO2'] <= 400) and (1001 <= row['CO'] <= 10) and (601 <= row['O3'] <= 240):
        return 'Buruk'
    else:
        return 'Sangat Buruk'

data['Kualitas_Udara'] = data.apply(classify_air_quality, axis=1)
data.head(100)
```

```python
import plotly.express as px

# Data contoh
df = pd.DataFrame({
    'year': [2013, 2013, 2013, 2014, 2014],
    'month': ['Maret', 'April', 'Mei', 'Maret', 'April'],
    'Kualitas_Udara': ['Baik', 'Tidak Baik', 'Baik', 'Baik', 'Tidak Baik']
})

# Menghitung jumlah kejadian per kategori
quality_counts = df['Kualitas_Udara'].value_counts()

# Membuat bar chart dengan Plotly
fig = px.bar(quality_counts, x=quality_counts.index, y=quality_counts.values,
             labels={'x': 'Kualitas Udara', 'y': 'Jumlah Kejadian'},
             title='Distribusi Kualitas Udara')
fig.show()
```
