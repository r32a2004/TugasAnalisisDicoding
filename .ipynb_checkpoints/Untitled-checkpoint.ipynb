### Cleaning Data

```python
medianPm=data['PM2.5'].median()
data['PM2.5'].fillna(medianPm)

medianPm10=data['PM10'].median()
data['PM10'].fillna(medianPm10)
medianSO2=data['SO2'].median()
data['SO2'].fillna(medianSO2)
medianN02=data['NO2'].median()
data['NO2'].fillna(medianN02)
medianCO=data['CO'].median()
data['CO'].fillna(medianCO)
median03=data['O3'].median()
data['O3'].fillna(median03)
data=data.dropna(axis=0)
```

Insight:

disni kita akan menganti nilai yang kosong dengan nilai median untuk kolom data yang jumlah nilai ilang nya bayak ,dengan mengunakan fungsi agregat yaitu median() untuk mengambil median ,lalu pengecekan apaka nilai data pada kolom tertentu kosong dengan fillna() jika iya maka nilai yang kosong atau ilang akan diganti dengan parameter isna , untuk kasus ini adalah median

untuk kolom data yang jumlah sedikit kita hapus mengunkan dropna

```python
data.isna().sum()
```

**Insight:**
- setelah melakukan pembersihan makan dapat di lihat data nilai hilang gak ada

```python
bulan_mapping = {
    1: "Januari", 2: "Februari", 3: "Maret", 4: "April",
    5: "Mei", 6: "Juni", 7: "Juli", 8: "Agustus",
    9: "September", 10: "Oktober", 11: "November", 12: "Desember"
}
data['month'] = data['month'].replace(bulan_mapping)

ganti_nama_arah_angin = {
    'N': 'North',
    'NNE': 'North-North-East',
    'NE': 'North-East',
    'ENE': 'East-North-East',
    'E': 'East',
    'ESE': 'East-South-East',
    'SE': 'South-East',
    'SSE': 'South-South-East',
    'S': 'South',
    'SSW': 'South-South-West',
    'SW': 'South-West',
    'WSW': 'West-South-West',
    'W': 'West',
    'WNW': 'West-North-West',
    'NW': 'North-West',
    'NNW': 'North-North-West'
}

# Terapkan mapping ke kolom 'wd'
data['wd'] = data['wd'].replace(ganti_nama_arah_angin )
```

**Insight:**
- selanjut nya kita menganti niali di dalam kolom data agar mempemudah untk analisis. kita ganti data dengan  sesuai bulan untuk mounth ,untuk data arah angin kita ganti dengan data sesunguh nya tidak megunakan singkatan
- hal itu bisa di lakukan dengan dua tahanp yaitu siapkan mapping untuk menganti nilai setiap data nya lalu gunakan fungsi replace dengan parameter mapping yang udah di buat

```python
data['CO']=data['CO']/1000
```

**Insight:**
- selanjutnya kita ubah format nilai co dari format µg/m³ ke  mg/m³ hal di lakukan karena biasa nya co di ukur dengan format  mg/m³. untuk melakukan konversi kita cukup menmbagi setiap data pada kolom co dengan 1000

## Visualization & Explanatory Analysis

### Pertanyaan 1 : Jumlah partikel kecil  bedarsakan  setiap bulan di satu tahun nya

```python
df = grupPM25ByMonth.reset_index()
heatmap_data = df.pivot_table(index='year', columns='month', values='PM2.5', aggfunc='sum')

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=.5)
plt.title('Heatmap Jumlah Partikel PM2.5 Bulan bedasrkan setiap Tahun')
plt.xlabel('Bulan')
plt.ylabel('Tahun')
plt.show()
```

**Insight:**

- visualisasi yang di tampilakan menguanakn seabord yaitu heatmap yang dimana heatmap itu mengkofersi dataframe jadi pivot tabel yang terdi dari x axis ataui indexs dan y axis yaitu colum nya, dan memilki nilai yang yang di kelompokan bedasrkan x dan y ,yang di mana untuk kasus ini saya menerapakan sum untuk data yang memilki kategori di setaiap bulan pada tahun tertentu. ini mirip banget dengan groupby() tapi kalo groupby() dia akan menghasilkan index dengan niali nya misal index nya itu adalah year dan month maka nilai nya adalah nilai yang ada dalam kategori itu ,dan nilai tergatung operasi agregat apa yang di lakukan,

- dan pada hetamap ini memilki intitas warna sesuai jumlah data nya untuk visualisasi ini di mulai dari kuning hingga biru tua

### Pertayaan6  visualisasi jumlah CO maximal dan minimal di setiap tahun nya

```python
import matplotlib.pyplot as plt

# Plot untuk CO Maksimum per Tahun
plt.figure(figsize=(12, 6))
bars_max = plt.bar(max_coYear.index, max_coYear.values, color='red', alpha=0.7, label='Maksimum CO')

plt.title('CO Maksimum per Tahun')
plt.xlabel('Tahun')
plt.ylabel('CO')
plt.legend()
plt.grid(True)

# Menampilkan nilai di atas batang
for bar in bars_max:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 1), va='bottom')  # va: vertical alignment

plt.show()

# Plot untuk CO Minimum per Tahun
plt.figure(figsize=(12, 6))
bars_min = plt.bar(min_coYear.index, min_coYear.values, color='blue', alpha=0.7, label='Minimum CO')

plt.title('CO Minimum per Tahun')
plt.xlabel('Tahun')
plt.ylabel('CO')
plt.legend()
plt.grid(True)

# Menampilkan nilai di atas batang
for bar in bars_min:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, round(yval, 1), va='bottom')

plt.show()

```

**Insight:**
- visualisasi ini dengan diagram batang yang dimana bar merah adala nilai maximun dari co setiap tahun  dan bar biru adalah nilai minum untuk co setiap tahun nya ,dan untuk pembuatan kita fokus pada fungsi plt.bar() yang digunakan membuat bar ,yang dimana memilki parameter inti terdiri dari x index untuk menunjukan ferkuensi data kalo y axis di gunakan untuk menampilkan pada rentang tertentu dan ada parameter warna untu memberi warna bar
- dan ada sebuah perlopingan untuk mengambil nilai asli dari dataset yang akan di taruh di atas diagram

###Klasifikasi Manual Untuk menetukan Udara Baik Atau Ga

```python

def classify_air_quality(row):
    if (row['PM2.5'] <= 12) and (row['PM10'] <= 50) and (row['SO2'] <= 20) and (row['NO2'] <= 40) and (row['CO'] <= 1) and (row['O3'] <= 100):
        return ' Baik'

    elif (row['PM2.5'] <= 35.4) and (row['PM10'] <= 100) and (81 <= row['SO2'] <= 80) and (row['NO2'] <= 200) and ( row['CO'] <= 3) and (row['O3'] <= 180):
        return 'Sedang'
    elif (row['PM2.5'] <= 55.4) and (row['PM10'] <= 150) and (381 <= row['SO2'] <= 380) or (row['NO2'] <= 400) and (1001 <= row['CO'] <= 10) and (601 <= row['O3'] <= 240):
        return 'Buruk'
    else:
        return 'Sangat Buruk'

data['Kualitas_Udara'] = data.apply(classify_air_quality, axis=1)
data.head(100)
```

**Insight:**
Disni kita melakukan klasifikasi apakah udara baik atau tidak melalui variabel PM2.5	PM10	SO2	NO2	CO	O3

berikut rentang nilai nya
kategori kualitas udara berdasarkan nilai PM2.5:

Baik: ≤ 12 µg/m³
Sedang: 12.1 - 35.4 µg/m³
Buruk: 35.5 - 55.4 µg/m³

PM10 berdasarkan rentang kualitas udara, berikut adalah kategorinya:

Baik: ≤ 50 µg/m³
Sedang: 51 - 100 µg/m³
Buruk: 101 - 150 µg/m³
Sangat Buruk: > 150 µg/m³

Berikut adalah rentang kualitas udara untuk SO₂ dalam µg/m³ yang biasanya digunakan:

Baik: ≤ 20 µg/m³
Sedang: 21 - 80 µg/m³
Buruk: 81 - 380 µg/m³
Sangat Buruk: > 380 µg/m³

Untuk NO₂ (Nitrogen Dioksida) dengan satuan µg/m³, rentang kualitas udara biasanya diatur sebagai berikut:

Baik: ≤ 40 µg/m³
Sedang: 41 - 200 µg/m³
Buruk: 201 - 400 µg/m³
Sangat Buruk: > 400 µg/m³


Untuk CO, rentang kualitas udara umumnya dalam format mg/m³ adalah sebagai berikut:

Baik: ≤ 1 mg/m³
Sedang: 1.1 - 3 mg/m³
Buruk: 3.1 - 10 mg/m³
Sangat Buruk: > 10 mg/m³

untuk O3
Baik: ≤ 100 µg/m³
Sedang: 101 - 180 µg/m³
Buruk: 181 - 240 µg/m³
Sangat Buruk: > 240 µg/m³

```python
count_baik_per_bulan = data[data['Kualitas_Udara'] == ' Baik'].groupby(['year', 'month']).size().reset_index(name='Jumlah_Baik')

heatmap_data = count_baik_per_bulan.pivot_table(index='year', columns='month', values='Jumlah_Baik')

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='g', linewidths=.5)
plt.title('Heatmap Jumlah udara yang bersih ')
plt.xlabel('Bulan')
plt.ylabel('Tahun')
plt.show()
print(count_baik_per_bulan)
```
