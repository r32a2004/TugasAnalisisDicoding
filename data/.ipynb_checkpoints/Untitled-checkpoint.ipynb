```python
import pandas as pd

# Memuat data
data = pd.read_csv('data.csv')

# Pastikan kolom tahun, bulan, dan hari adalah integer
data['year'] = data['year'].astype(int)
data['month'] = data['month'].astype(int)
data['day'] = data['day'].astype(int)

# Membuat DataFrame dengan kolom year, month, dan day
date_df = pd.DataFrame({
    'year': data['year'],
    'month': data['month'],
    'day': data['day']
})

# Menggabungkan kolom tahun, bulan, dan hari menjadi satu kolom datetime
data['datetime'] = pd.to_datetime(date_df)

# Mengatur kolom datetime sebagai index (opsional, mempermudah analisis)
data.set_index('datetime', inplace=True)

# Mengelompokkan data berdasarkan tahun
annual_groups = data.groupby(data.index.year)

# Menghitung korelasi tahunan antara CO, PM2.5, dan PM10
def calculate_annual_correlation(group):
    return group[['CO', 'PM2.5', 'PM10']].corr().loc['CO', ['PM2.5', 'PM10']]

# Menghitung korelasi untuk setiap tahun
annual_correlation = annual_groups.apply(calculate_annual_correlation)

# Menampilkan hasil
print(annual_correlation)
```

```python
import pandas as pd

# Memuat data
data = pd.read_csv('data.csv')

# Menampilkan beberapa baris pertama data untuk verifikasi
print(data.head())
# Mengelompokkan data berdasarkan arah mata angin dan menghitung rata-rata konsentrasi CO
wind_direction_groups = data.groupby('wd')['CO'].mean()

# Menampilkan hasil
print(wind_direction_groups)
# Menghitung statistik deskriptif untuk konsentrasi CO berdasarkan arah mata angin
wind_direction_stats = data.groupby('wd')['CO'].describe()

# Menampilkan statistik deskriptif
print(wind_direction_stats)
import matplotlib.pyplot as plt

# Plot rata-rata konsentrasi CO berdasarkan arah mata angin
wind_direction_groups.plot(kind='bar')
plt.xlabel('Arah Mata Angin')
plt.ylabel('Rata-rata Konsentrasi CO')
plt.title('Rata-rata Konsentrasi CO Berdasarkan Arah Mata Angin')
plt.show()

```
