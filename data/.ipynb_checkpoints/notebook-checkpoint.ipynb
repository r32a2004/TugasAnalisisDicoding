# Air Quality Data Analysis Project

```python
# Importing libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

## 1. Data Wrangling

### 1.1 Gathering Data

```python
# Directory containing the CSV files
data_dir = "/content/data/"

# Dictionary to store DataFrames
dfs = {}

# Load CSV files into DataFrames
for file_name in os.listdir(data_dir):
    df_name = os.path.splitext(file_name)[0].split("_")[2]  # Use file name without extension as DataFrame name
    dfs[df_name] = pd.read_csv(os.path.join(data_dir, file_name)).drop(['No'], axis = 1)
```

```python
dfs['Wanshouxigong'].head()
```

### 1.2 Assessing Data

1.2.1 Simplify variables

```python
# Define a function to combine year, month, day, and hour into a datetime column
def combine_datetime(df):
    df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
    df.drop(['year', 'month', 'day', 'hour'], axis=1, inplace=True)
    return df

# Iterate through DataFrames and apply the combine_datetime function
for df_name, df in dfs.items():
    dfs[df_name] = combine_datetime(df)
```

1.2.2 Check for Duplicates

```python
# Initialize a dictionary to store duplicate counts and total sample counts
duplicate_counts = {}
total_sample_counts = {}

# Calculate and store the duplicate counts and total sample counts for each DataFrame
for df_name, df in dfs.items():
    duplicate_counts[df_name] = df.duplicated().sum()
    total_sample_counts[df_name] = len(df)  # Calculate the total number of samples

# Create DataFrames from the dictionaries
duplicate_counts_df = pd.DataFrame.from_dict(duplicate_counts, orient='index', columns=['Duplicate Count'])
total_sample_counts_df = pd.DataFrame.from_dict(total_sample_counts, orient='index', columns=['Total Sample Count'])

# Combine the two DataFrames by concatenating them horizontally
pd.concat([total_sample_counts_df, duplicate_counts_df], axis=1)
```

1.2.3 Check for All of The Data Types

```python
# Initialize a dictionary to store data types for each DataFrame
dtype_dict = {}

# Iterate through DataFrames and collect data types
for df_name, df in dfs.items():
    dtype_dict[df_name] = df.dtypes

# Create a DataFrame from the dtype_dict and transpose it
# to have DataFrames as rows and columns as data types
pd.DataFrame(dtype_dict).transpose()
```

1.2.4 Check for Missing Values

```python
# Initialize a dictionary to store null value counts for each DataFrame
null_counts = {}

# Calculate and store the null value counts for each DataFrame
for df_name, df in dfs.items():
    null_counts[df_name] = df.isna().sum()

# Create a DataFrame from the null_counts dict and transpose it
pd.DataFrame(null_counts).transpose()
```

### 1.3 Data Cleaning

1.3.1 Handle the numerical missing values with the median value

```python
# Calculate and store the median values for each numerical column
median_values = {}

for df_name, df in dfs.items():
    # Exclude non-numeric columns
    numeric_columns = df.select_dtypes(include=['number']).columns.difference(['No'])

    # Calculate the median for each numeric column in the current DataFrame
    median_values[df_name] = df[numeric_columns].median()

# Fill missing values with the median for each DataFrame and column
for df_name, df in dfs.items():
    # Exclude non-numeric columns
    numeric_columns = df.select_dtypes(include=['number']).columns.difference(['No'])

    # Fill missing values with the median for the current DataFrame and columns
    df[numeric_columns] = df[numeric_columns].fillna(median_values[df_name])
```

1.3.2 Handle the categorical missing values with their surrounding value

```python
# Loop through the DataFrames and fill missing values with forward fill
for df_name, df in dfs.items():
    df['wd'].fillna(method='ffill', inplace=True)
```

1.3.3 Double check if there are any NaN values remaining

```python
# Initialize a dictionary to store null value counts for each DataFrame
null_counts = {}

# Calculate and store the null value counts for each DataFrame
for df_name, df in dfs.items():
    null_counts[df_name] = df.isna().sum()

# Create a DataFrame from the null_counts dict and transpose it
pd.DataFrame(null_counts).transpose()
```

## 2. Exploratory Data Analysis (EDA)

### 2.1 Explore the Mean Values of the Parameters

```python
# Create a dictionary to store mean values for each parameter
mean_values = {}

# Loop through the DataFrames in the dfs dictionary
for df_name, df in dfs.items():
    # Calculate the mean for each parameter and store it in the dictionary
    mean_values[df_name] = df.mean(numeric_only=True)

# Create a DataFrame from the dictionary of mean values
pd.DataFrame(mean_values).transpose()
```

### 2.2 Explore the Median Values of the Parameters

```python
# Create a dictionary to store median values for each parameter
median_values = {}

# Loop through the DataFrames in the dfs dictionary
for df_name, df in dfs.items():
    # Calculate the median for each parameter and store it in the dictionary
    median_values[df_name] = df.median(numeric_only=True)

# Create a DataFrame from the dictionary of median values
pd.DataFrame(median_values).transpose()
```

### 2.3 Combine all of the DataFrames in the dfs dictionary into one DataFrame

```python
# Concatenate all DataFrames into one DataFrame along the rows
combined_df = pd.concat(list(dfs.values()), axis=0)

# Reset the index of the combined DataFrame
combined_df.reset_index(drop=True, inplace=True)
```

### 2.4 Explore the combined DataFrame

```python
combined_df.describe()
```

### 2.5 Add a category column to the DataFrame based on the worst category among the six parameters

```python
# Define the ranges and AQI (Air Quality Index) upper bounds
category_ranges = [
    'Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous'
]
pm25_ranges = [12, 35.4, 150.4, 250.4, float('inf')]
pm10_ranges = [54, 154, 254, 354, 424, float('inf')]
so2_ranges = [35, 75, 185, 304, 604, float('inf')]
no2_ranges = [53, 100, 360, 649, 1249, float('inf')]
co_ranges = [4400, 9400, 12400, 15400, 30400, float('inf')]
o3_ranges = [54, 70, 85, 105, 200, float('inf')]

# Define a function to categorize a value based on the given ranges
def categorize(value, ranges):
    for max_val in ranges:
        if value <= max_val:
            return ranges.index(max_val)

# Create a new column 'Category' by applying the categorize function to each row
combined_df['Category'] = combined_df.apply(lambda row: category_ranges[max(
    categorize(row['PM2.5'], pm25_ranges),
    categorize(row['PM10'], pm10_ranges),
    categorize(row['SO2'], so2_ranges),
    categorize(row['NO2'], no2_ranges),
    categorize(row['CO'], co_ranges),
    categorize(row['O3'], o3_ranges),
)], axis=1)
```

## 3. Visualization & Explanatory Analysis

### 3.1 Are there significant correlations between different pollutants or with the weather conditions?

```python
correlation_df = combined_df.corr(numeric_only=True)

# Create a mask to hide the upper triangular part of the table
mask = np.triu(np.ones_like(correlation_df, dtype=bool))

# Create a custom color map for shading
cmap = sns.diverging_palette(240, 10, as_cmap=True)

# Create a heatmap with shading
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_df, annot=True, fmt=".2f", cmap=cmap, mask=mask)

plt.title("Correlation Heatmap")
plt.show()
```

The heatmap shows the correlations between various air quality and weather variables. Here are some insights that can be derived from the correlations:

1.   **PM2.5** and **PM10** are **highly correlated**. This is expected, as both PM2.5 and PM10 measure particulate matter of different sizes.

2.   High levels of **PM2.5** are associated with high levels of **SO2**, **NO2**, and **CO**, which can be indicative of poor air quality.

3.   Higher **ozone** levels are associated with lower levels of **PM2.5**, **PM10**, **SO2**, **NO2**, and **CO**. Ozone is often involved in chemical reactions that reduce the concentration of these pollutants.

4.   High temperatures are associated with higher dew points.

5.   Higher **temperatures** may be associated with lower levels of **SO2**, **NO2**, **CO**, and **PM2.5**, possibly due to increased dispersion and chemical reactions.

6.   As pressure decreases, temperature tends to increase, and dew point tends to decrease.

7.   Higher **wind speeds** (WSPM) are associated with lower levels of **PM2.5**, **PM10**, **SO2**, **NO2**, and **CO**, which can help disperse air pollutants.

8.   RAIN has weak correlations with most variables, indicating that rainfall may not have a strong influence on these air quality and weather variables.

These insights provide a better understanding of how different air quality and weather parameters are related to each other, which can be valuable for assessing air quality and making predictions. For example, the strong correlation between PM2.5 and PM10 suggests that these measurements can be used interchangeably in some cases, while the negative correlation between ozone and other pollutants may indicate the effectiveness of ozone in reducing pollution levels through chemical reactions.

The stacked bar chart represent air quality index (AQI) categories for different monitoring stations in a specific region. AQI is typically used to assess air quality and its impact on health. Here are some insights that can be generated from this data:

1.   The station with the highest count in the "**Good**" air quality category is **Changping** with 2238 counts, followed by **Shunyi** with 2044 counts. These stations seem to have more days with better air quality compared to others.

2.   **Shunyi** has the highest count in the "**Very Unhealthy**" category with 4983 counts, followed by **Guanyuan** with 5639 counts. This suggests that Shunyi frequently experiences days with very poor air quality.

3.   The stations with the highest counts in the "**Moderate**" air quality category are **Aotizhongxin** and **Dingling**, both with counts above 7000. This indicates that these stations often have moderate air quality conditions.

4.   **Guanyuan**, **Dongsi**, and **Wanliu** have relatively high counts in the "**Unhealthy**" and "**Unhealthy for Sensitive Groups**" categories. This suggests that people with respiratory conditions in these areas may be at risk on many days.

5.   **Nongzhanguan** and **Tiantan** have relatively **balanced** counts across the different air quality categories. This indicates that the air quality in these areas tends to vary between different categories.

6.   **Gucheng** has a relatively high count in the "**Hazardous**" category, indicating that this station frequently experiences hazardous air quality conditions, which can be harmful to everyone.

7.   **Huairou** stands out with a relatively low count in the "**Very Unhealthy**" category compared to other stations, suggesting that very poor air quality is less common in this area.

The region covered by these monitoring stations seems to experience a wide range of air quality conditions, with episodes of poor air quality that fall into the "Hazardous" and "Very Unhealthy" categories. This data could be useful for understanding the health risks associated with air pollution in this area and for making decisions regarding public health and environmental policies.

### 3.3 Can we identify potential sources of pollution based on wind direction?

```python
# Define the air quality parameters
parameters = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']
```

```python
# Calculate the median for each parameter by 'wd' (wind direction)
wd_table = combined_df[parameters].groupby(combined_df['wd']).median()

# Normalize each column to its maximum value
normalized_wd_table = wd_table / wd_table.max()

# Create a heatmap with the transposed data
plt.figure(figsize=(10, 3))
sns.heatmap(normalized_wd_table.T, cmap="YlGnBu", annot=True, fmt=".2f", linewidths=0.5)
plt.title("Normalized Median Values by WD and Parameter")
plt.xlabel("WD")
plt.ylabel("Parameter")
plt.show()
```

The heatmap that provides median values of various parameters by wind direction offers several insights:

1.   **PM2.5**, **PM10**, **SO2**, **NO2**, and **CO** levels tend to be **higher for wind directions from the east** (E, ENE, ESE), **southeast** (S, SE, SSE), and **southwest** (SSW, SW, W) while they are **lower for wind directions from the northwest** (NW, NNW) and **west-northwest** (WNW). This suggests that the source of these pollutanrs might be associated with certain wind patterns.

2.   **O3** levels are **higher for wind directions from the south-southwest** (SSE, SSW, SW). This could indicate differences in pollution sources and atmospheric processes.

Overall, this heatmap provides a summary of how various environmental parameters relate to wind direction, which can be useful for understanding local air quality patterns and potentially identifying sources of pollution based on prevailing wind patterns.

### 3.4 Are there seasonal patterns or fluctuations in air quality data?

```python
# Create a 3x2 grid of subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 10))
fig.suptitle('Distributions of Air Quality Parameters Over Time', fontsize=16)

# Loop through the parameters and create subplots
for i, parameter in enumerate(parameters):
    row = i // 2  # Row index (0, 0, 1, 1, 2, 2)
    col = i % 2   # Column index (0, 1, 0, 1, 0, 1)

    # Plot the distribution of the parameter
    axes[row, col].plot(combined_df['datetime'], combined_df[parameter], linestyle='-')
    axes[row, col].set_title(f'{parameter} Distribution')
    axes[row, col].set_xlabel('Datetime')
    axes[row, col].set_ylabel(parameter)
    axes[row, col].grid(True)
    axes[row, col].tick_params(axis='x', rotation=45)

# Adjust the layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the subplots
plt.show()
```

```python
# Calculate the median for each parameter by month
median_table = combined_df.groupby(pd.to_datetime(combined_df['datetime']).dt.month).median(numeric_only=True)

# Normalize each column to its maximum value
normalized_median_table = median_table / median_table.max()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(normalized_median_table, cmap="YlGnBu", annot=True, fmt=".2f", linewidths=0.5)
plt.title("Normalized Median Values by Parameter and Month")
plt.xlabel("Parameter")
plt.ylabel("Month")

# Adjust the layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

From the distributions of air quality parameters over time, several insights can be derived:

1.   **PM2.5**, **PM10**, **SO2**, and **CO** tend to be **higher in the winter months** (e.g., November to February) compared to the rest of the year. This suggests that air pollution, particularly particulate matter and certain gases, tends to be worse during the colder months.

2.   O3 (**Ozone**) levels **peak during the summer months** (e.g., June to August), indicating a seasonal variation in ozone concentration.

3.   The seasonal variation in temperature is expected and contributes to the observed patterns in pollutant concentrations.

4.   Rainfall does not have a significant influence on air quality during most of the year.

5.   Wind speed does not exhibit strong seasonal patterns.

6.   Pollutants like **PM2.5**, **PM10**, **SO2**, **NO2**, and **CO** show variations in concentration throughout the year, which may be influenced by factors such as weather conditions, emissions, and human activities.

7.   Certain pollutants (e.g., **PM2.5**, **PM10**) tend to decrease as temperatures rise.

### 3.5 How do air quality measurements vary between different monitoring stations?

```python
# Create a 3x2 grid of subplots with vertical spacing (hspace)
fig, axes = plt.subplots(3, 2, figsize=(15, 12))
fig.suptitle('Distributions of Air Quality Parameters by Station', fontsize=16)

# Loop through the parameters and create boxplots for each one
for i, parameter in enumerate(parameters):
    row = i // 2  # Row index (0, 0, 1, 1, 2, 2)
    col = i % 2   # Column index (0, 1, 0, 1, 0, 1)

    # Create a boxplot for the parameter
    sns.boxplot(x='station', y=parameter, data=combined_df, ax=axes[row, col])
    axes[row, col].set_title(f'{parameter} Distribution')
    axes[row, col].set_xlabel('Station')
    axes[row, col].set_ylabel(parameter)

# Rotate x-axis labels for better readability
for ax in axes.flat:
    ax.tick_params(axis='x', rotation=45)

# Adjust the layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the subplots
plt.show()
```

```python
# Convert 'datetime' column to datetime objects
combined_df['datetime'] = pd.to_datetime(combined_df['datetime'])

# Create a 3x2 grid of subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 10))
fig.suptitle('Distributions of Air Quality Parameters Over Time', fontsize=16)

# Create an empty list to collect legend handles and labels
legend_handles = []

# Loop through the parameters and create subplots
for i, parameter in enumerate(parameters):
    row = i // 2  # Row index (0, 0, 1, 1, 2, 2)
    col = i % 2   # Column index (0, 1, 0, 1, 0, 1)

    # Group data by month and station, then plot the distribution of the parameter for each station
    grouped_data = combined_df.groupby([combined_df['datetime'].dt.to_period("M"), 'station'])[parameter].mean().unstack()
    monthly_index = grouped_data.index.to_timestamp()  # Convert the index to a datetime series
    for station in grouped_data.columns:
        line, = axes[row, col].plot(monthly_index, grouped_data[station], linestyle='-', label=station)
        if row == 0 and col == 0:
            legend_handles.append(line)

    axes[row, col].set_title(f'{parameter} Distribution')
    axes[row, col].set_xlabel('Month')
    axes[row, col].set_ylabel(parameter)
    axes[row, col].grid(True)
    axes[row, col].tick_params(axis='x', rotation=45)

    # Create a single legend outside of the subplots
    if row == 0 and col == 1:
        axes[row, col].legend(handles=legend_handles, loc='best', bbox_to_anchor=(1, 1), title='Stations')

# Adjust the layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the subplots
plt.show()
```

From the distributions of air quality parameters by station, several insights can be derived:

1.   **Shunyi** and **Wanliu** have **higher** **PM2.5** and **PM10** levels, suggesting potentially worse air quality in terms of particulate matter.

2.   **SO2** levels are **relatively low** across all locations, indicating that SO2 pollution is generally not a significant concern in these areas.

3.   **Aotizhongxin**, **Wanliu**, and **Wanshouxigong** have **higher NO2 levels**.

4.   **Tiantan** and **Wanshouxigong** have **higher CO levels**.

5.   **Dingling** and **Huairou** have **higher O3 levels**.
